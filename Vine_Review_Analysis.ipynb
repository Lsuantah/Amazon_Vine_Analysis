```python
from pandas import pandas as pd
```

```python
df = pd.read_csv("C:/Users/User/Desktop/WorkSpace/HomeWork/Challenge/Resources/vine_table.csv",header=0)
df.head()
```

```python
#retrieve all the rows where the total_votes count is equal to or greater than 20 
df_vote = df[df.total_votes >= 20]
df_vote.head()
```

```python
#create a new DataFrame or table to retrieve all the rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%

df_rows = df_vote[(df_vote.helpful_votes/df_vote.total_votes)>0.5]
df_rows
```

```python
#retrieves all the rows where a review was written as part of the Vine program (paid), vine == 'Y'
df_filter_Y = df_rows[df_rows.vine=='Y']
df_filter_Y
```

```python
df_filter_N = df_rows[df_rows.vine=='N']
df_filter_N
```

```python
#Determine the total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews for the two types of review

TotalReviews_Y = df_filter_Y.review_id.count()
FiveStarReviews_Y = df_filter_Y.review_id[df_filter_Y.star_rating == 5.0].count()
FiveStarPct_Y = FiveStarReviews_Y/TotalReviews_Y

TotalReviews_N = df_filter_N.review_id.count()
FiveStarReviews_N = df_filter_N.review_id[df_filter_N.star_rating == 5.0].count()
FiveStarPct_N = FiveStarReviews_N/TotalReviews_N
```

```python
df_analysis = pd.DataFrame([{'Vine flag':'Y','Total Reviews':TotalReviews_Y,'Five Star Reviews':FiveStarReviews_Y,'Five Star Reviews %':FiveStarPct_Y},
    {'Vine flag':'N','Total Reviews':TotalReviews_N,'Five Star Reviews':FiveStarReviews_N,'Five Star Reviews %':FiveStarPct_N}
    ])
df_analysis.set_index('Vine flag',inplace=True)
```
